```python
'''
3.1.1. a
Observing outputs of SCC-PI by IE model
''' 
#

pii_tid_list = [6,7,8,25,30,41,45,47,56,92]

pred_tuples = []        
for ind_li in pii_tid_list:
    pii_tidx = comp_list[ind_li]
    if pii_tidx in output['identifier']:
        ind = output['identifier'].index(pii_tidx)
        print(pii_tidx)
        pred_tuples += output['pred_tuples'][ind]
        print(output['pred_tuples'][ind])
        print('------------------------')
    else:
        print(f'Not found pii = {pii_tidx}')
        
'''
Output analysis:
6,7,8,25, 56 - No composition extr
30, 41, 47, 92 - corr extr
45 - incorrect

As in 60% of the case the desired composition hasnt been extracted and also Material F1 is 47.19% only,
we can say that the challenge is valid.
'''
```

```python
pickle.dump(pred_tuples, open('pred_tuples_for_scc_pi.pkl', 'wb'))
```
